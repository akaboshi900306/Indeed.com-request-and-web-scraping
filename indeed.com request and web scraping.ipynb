```python
def getTitle_name(soup):
    title =[]
    for k in soup.find_all("h2", class_="title"):
        try:
            a= k.find("a",attrs={"class":"jobtitle turnstileLink"})["title"]
            title.append(a)
        except:
            title.append("None")
    return(title)
```

```python
title = getTitle_name(soup)
```

```python
def getCompany_name(soup):
    company = []
    for div in soup.find_all(name="div", attrs={"class":"row"}):
        try:
            a = div.find(name="span", attrs={"class":"company"}).text.strip()
            company.append(a)
        except: 
            company.append("None")
    return(company)

    
```

```python
company =getCompany_name(soup)
```

```python
def getLocation_name(soup):
    location= []
    for div in soup.find_all(name="div", attrs={"class":"row"}):
        try:
            a = div.find(name="div", attrs={"class":"location accessible-contrast-color-location"}).text
            location.append(a)
        except: 
            location.append("None")
    return(location)
```

```python
location = getLocation_name(soup)
```

```python
def getSalary(soup):
    salary = []
    for k in soup.find_all(name="div", attrs={"class":"row"}):
        try:
            a =k.find("span", class_="salaryText").text.strip()
            salary.append(a)
        except:
            salary.append("Nothing")
    return(salary)
```

```python
salary = getSalary(soup)
```

```python
def getSummary(soup):
    summaries = []
    spans = soup.find_all("div", attrs={"class": "summary"})
    for span in spans:
        try:
            span = span.text.strip()
            span = span.replace("\n","")
            summaries.append(span)
        except:
            summaries.append(span.text.strip())
    return(summaries)
    
summary = getSummary(soup)
```

```python
df = pd.DataFrame({"Title":title, "Company":company, "Location":location, "Salary":salary, "Summary":summary})
```
